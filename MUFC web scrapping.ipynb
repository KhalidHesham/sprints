# Web Scraping Project 
## Scraping [Sports Reference](https://fbref.com/en/squads/19538871/2022-2023/matchlogs/c9/schedule/Manchester-United-Scores-and-Fixtures-Premier-League) available data for Manchester United PL run in the 22/23 season


```python
import pandas as pd

# Send a GET request to the URL and extract HTML content
url = "https://fbref.com/en/squads/19538871/2022-2023/matchlogs/c9/schedule/Manchester-United-Scores-and-Fixtures-Premier-League"
html_content = pd.read_html(url)

# Assuming the desired table is the first one, you can access it as follows
table_data = html_content[0]

# Display the table data
print(table_data)
```

```python
df = pd.DataFrame(table_data)
# Display the table data
df.head()
```

```python
shape  = df.shape
print(shape) 
```

```python
df.drop(["Match Report","Notes"], axis = 1 , inplace = True)
```

```python
df.head()
```

```python
df.drop(["Attendance"], axis = 1 , inplace = True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
x = df['Venue']
y = df['GF']

# Create a bar plot using Seaborn
sns.barplot(x=x, y=y)

# Add labels and title
plt.xlabel('Venue')
plt.ylabel('GF')
plt.title('Relation between Goals For Home and Away')

# Display the graph
plt.show()
```
